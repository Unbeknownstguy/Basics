## Comparison

```python
x1 = [2, 5, 6, 7, 8]
y1 = [3, 7, 4, 12, 5]

x2 = [4, 5, 8, 9, 8]
y2 = [3, 5, 7, 12, 6]
```

```python
plt.plot(x1,y1,color='g', linewidth=6)
plt.plot(x2,y2,color='r', linewidth=4)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Comparison Graph')
plt.show()
```

```python
plt.plot(x1,y1,color='g', linewidth=6)
plt.plot(x2,y2,color='r', linewidth=4)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Comparison Graph')

plt.annotate(xy=[2, 3], text= 'First line')
plt.annotate(xy=[4, 4], text= 'Second line')

plt.legend(['First _line', 'Second_line'], loc=4)

plt.show()
```

# Vertical Bar Graph

```python
x1 = [2, 5, 6, 7, 8]
y1 = [3, 7, 4, 12, 5]

x2 = [4, 5, 8, 9, 8]
y2 = [3, 5, 7, 12, 6]

plt.bar(x1,y1,color='g', linewidth=6)
plt.bar(x2,y2,color='r', linewidth=4)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Comparison Graph')

plt.legend(['First _line', 'Second_line'], loc=4)
plt.show() 
```

## Horizontal Bar Graph

```python
x1 = [2, 5, 6, 7, 8]
y1 = [3, 7, 4, 12, 5]

x2 = [4, 5, 8, 9, 8]
y2 = [3, 5, 7, 12, 6]

plt.barh(x1,y1,color='g', linewidth=6)
plt.barh(x2,y2,color='r', linewidth=4)

plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title('Comparison Graph')

plt.legend(['First _line', 'Second_line'], loc=4)
plt.show()
```

## Bubble Graph

```python
x = [2, 5, 6, 7, 8]
y = [3, 7, 4, 12, 5]
```

```python
plt.scatter(x,y, s=np.random.rand(5)*1000, c=np.random.rand(5))

plt.title("Bubble Graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```
