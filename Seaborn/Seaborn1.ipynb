It's almost impossible to derive meaning or insights by looking at the raw dataset,
but when we plot those data in a suitable graph or plots it helps us to give some meanings that we want. 
This is where Data Visualization comes into play.

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Seaborn has some built-in datasets

```python
# total bill vs tip dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
# setting a theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x = 'total_bill', y= 'tip', col='time', hue= 'smoker', style= 'smoker', size= 'size')
```

```python
# loding the iris dataset
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.shape
```

### Scatter Plot

```python
sns.scatterplot(x='sepal_length', y= 'petal_length', hue= 'species', data=iris)   # hue differentiates the species with different color
```

```python
sns.scatterplot(x='sepal_length', y='petal_width', hue='species', data=iris)
```

```python
# loading the titanic dataset
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

### Count Plot

```python
sns.countplot(x='class', data=titanic)
```

```python
sns.countplot(x='survived', data=titanic)
```

### Bar chart

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
# house price dataset
from sklearn.datasets import load_boston
house_boston = load_boston()

house = pd.DataFrame(house_boston.data, columns = house_boston.feature_names)
house['PRICE'] = house_boston.target
```

```python
print(house_boston)
```

```python
house.head()
```

```python
house.shape
```

### Distribution Plot

```python
sns.displot(house['PRICE'])
```

```python
sns.distplot(house['PRICE'])   # plots distribution curve
```

### Correlation
1. Positive Correlation
2. Negative Correlation

### Heat Map

```python
correlation = house.corr()
```

```python
# constructing a Heat Map
plt.figure(figsize=(10, 10))

# cbar --> color_bar
#fmt --> number of floating point
sns.heatmap(correlation, cbar=True, square=True, fmt = '.1f', annot=True, annot_kws={'size':8}, cmap='Blues')
```
